```python
import pandas as pd
%matplotlib inline
from pathlib import Path
csvpath = Path('../Resources/high_frequency_euro_ETF_bid_ask_spreads.csv')
df = pd.read_csv(csvpath, parse_dates=True, index_col='timestamp')
df.head()
```

```python
# Resample to 10 Seconds ('10S'), take the mean, and drop the NaN values
ten = df.resample('10S').mean().dropna()
```

```python
# Plot the resulting bid-ask spread data
ten.plot()
```

```python
# Calculate the autocorrelation
ten.bid_ask_spread.autocorr(lag=1)
```

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
```

```python
# Plot the ACF
plot_acf(ten.bid_ask_spread, lags=25, zero=False)
```

```python
# Plot the PACF
plot_pacf(ten.bid_ask_spread, lags=25, zero=False)
```
