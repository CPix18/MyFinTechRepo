```python
import pandas as pd
%matplotlib inline
```

```python
from pathlib import Path
csvpath = Path('../Resources/BTC5Y_Data.csv')
df = pd.read_csv(csvpath, parse_dates=True, index_col='Date')
df.head()
```

```python
df.Close.plot()
```

```python
# Differencing the Settle price stationarizes the time series
df.Close.diff().plot()
```

# Autocorrelation and Partial Autocorrelation Analysis

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
```

```python
# Plot the autocorrelation of the Settle Prices
plot_acf(df.Close, zero = False, lags = 50)
```

```python
# Plot the partial autocorrelation of the returns
plot_pacf(df.Close, zero = False, lags = 50)
```

# ARIMA Model

```python
from statsmodels.tsa.arima_model import ARIMA
```

```python
# Create an ARIMA model
model = ARIMA(df['Close'], order=(2, 1, 2))
```

```python
# Fit the model to the data
results = model.fit()
```

```python
# Print the model summary
results.summary()
```

```python
# Plot the Futures Price Forecast
pd.DataFrame(results.forecast(steps=500)[0]).plot(title="BTC Spot Forecast")
```
