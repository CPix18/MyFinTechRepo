# Instructor Do: Getting Data for Sentiment Analysis

```python
# Initial imports
import os
import pandas as pd
from newsapi import NewsApiClient

```

```python
# Reading the News API key enviroment variable
api_key = os.getenv("NEWSAPI_KEY")

```

```python
# Create a newsapi client
newsapi = NewsApiClient(api_key=api_key)

```

```python
# Fetch top news articles about oil
coronavirus_headlines = newsapi.get_top_headlines(q="coronavirus", language="en", country="us")
coronavirus_headlines

```

```python
# Print total articles
print(f"Total articles about coronavirus: {coronavirus_headlines['totalResults']}")

# Show sample article
coronavirus_headlines["articles"][0]

```

```python
# Transform the response dictionary to a DataFrame
coronavirus_df = pd.DataFrame.from_dict(coronavirus_headlines["articles"])

coronavirus_df.head()

```

```python
# Fetch all the news about Facebook Libra
libra_headlines = newsapi.get_everything(
    q="facebook libra", language="en", sort_by="relevancy"
)

# Print total articles
print(f"Total articles about Facebook Libra: {libra_headlines['totalResults']}")

# Show sample article
libra_headlines["articles"][0]

```

```python
# Transform the response dictionary to a DataFrame
libra_df = pd.DataFrame.from_dict(libra_headlines["articles"])

libra_df.head()
```
