### Instructor Demo: Indexing

This program reads performs several slice and dice operations through indexing via loc and iloc functions.

```python
# Import libraries and dependencies
import pandas as pd
from pathlib import Path
```

### Read CSV in as DataFrame

```python
# Set the file path
file_path = Path('../Resources/people.csv')

# Read in the CSV as a DataFrame
people_csv = pd.read_csv(file_path)
people_csv.head()
pc = people_csv
pc.head
```

```python
pc.iloc[:3, [2,3,6]]
```

```python
pc.shape
```

```python
smalldf = pc.iloc[ :, [2,3,6]]
smalldf["age"].value_counts()
```

```python
pc.head()
```

```python
pc.loc[pc["gender"] == 'F', "age"] = 22
pc.head()
```
