# Sorting Pandas DataFrames

## 1. Import Libraries and Dependencies

```python
# Import libraries and dependencies
import pandas as pd
from pathlib import Path
```

## 2. Create a Path to the File Using Pathlib

```python
# Use the Pathlib libary to set the path to the CSV
people_data = Path("../Resources/people_cleansed.csv")
```

## 3. Read the CSV into a Pandas DataFrame

```python
# Read in the CSV as a DataFrame
people_df = pd.read_csv(people_data)
people_df.head()
```

## 4. Sort DataFrame in Ascending Order

```python
# Sort ascending (default)
people_df.sort_values("Occupation")
```

## 5. Sort DataFrame in Descending Order

```python
# Sort descending
people_df.sort_values("Occupation", ascending=False)
```
